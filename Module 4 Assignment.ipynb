```python
import pandas as pd

# Load the dataset
df = pd.read_csv('netflix_data.csv')

# Check for missing values
print(df.isnull().sum())

# Fill missing values or drop rows
df['rating'].fillna('Unknown', inplace=True)  # Example: Filling missing ratings
df.dropna(subset=['title'], inplace=True)  # Dropping rows where title is missing
```

```python
# Describe the data
print(df.describe())

# Statistical analysis
print(df['release_year'].value_counts())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

import pandas as pd

df = pd.read_csv('netflix_data.csv')

# Plot the most watched genres
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='type', order=df['type'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Most Watched Genres')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.show()
```

```python
# Plot the distribution of ratings
plt.figure(figsize=(10, 6))
sns.histplot(df['rating'], bins=10, kde=True)
plt.title('Ratings Distribution')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```
